```python
# **Задание 1**

# Используем файл keywords.csv.
# Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'Центр'.
# Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
# geo_data = {
#      'Центр': ['москва', 'тула', 'ярославль'],
#      'Северо-Запад': ['петербург', 'псков', 'псков'],

#      'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
#  }

import pandas as pd
data = pd.read_csv('keywords.csv')

# создаем функцию с 3 вариантами регоинов и 1 вариантом по умолчанию
def geo_cl(row):
    if 'москва' in row['keyword'] or 'тула' in row['keyword'] or 'ярославль' in row['keyword']:
        return 'Центр'
    elif 'петербург' in row['keyword'] or 'псков' in row['keyword'] or 'псков' in row['keyword']:
        return 'Северо-Запад'
    elif 'владивосток' in row['keyword'] or 'сахалин' in row['keyword'] or 'хабаровск' in row['keyword']:
        return 'Дальний Восток'
    else:
        return 'undefined'

# прогоняем функцию с помощью апплай
data['region'] = data.apply(geo_cl, axis=1)

# проверяем, сколько значений, отличных от 'undefined', получилось в итоге
data['region'].value_counts()
```

```python
# **Задание 2**

# Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера.


# Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
#     - оценка 2 и меньше - низкий рейтинг
#     - оценка 4 и меньше - средний рейтинг
#     - оценка 4.5 и 5 - высокий рейтинг

# Результат классификации запишите в столбец class

import pandas as pd
movies_dt = pd.read_csv('movies.csv')
ratings_dt = pd.read_csv('ratings.csv')

# создаем общую таблицу Х
x = ratings_dt.merge(movies_dt, how='left')

# делаем таблицу индекс- название фильма - средний ретинг
films_rating = x.groupby('title').mean().reset_index()[['title', 'rating']]


# создаем функцию с вариантами оценки
def range_maker(row):
    if row['rating'] <= 2.0:
        return 'low'
    elif (row['rating'] > 2.0) and (row['rating'] <= 4.0):
        return 'medium'
    elif row['rating'] > 4.0:
        return 'high'
    else:
        return 'undefined'
    
# прогоняем функцию с помощью апплай
films_rating['class'] = films_rating.apply(range_maker, axis=1)


films_rating.head(100)
```

```python
# **Задание 3**

# Посчитайте среднее значение Lifetime киноманов (пользователи, которые поставили 100 и более рейтингов). Под Lifetime понимается разница между максимальным и минимальным значением timestamp для каждого пользователя. Ответ дайте в днях.

#начинаем традиционно
import pandas as pd

ratings_dt = pd.read_csv('ratings.csv')

# делаем как в лекции
grouped_dt = ratings_dt.groupby('userId').count().reset_index()
grouped_dt.head(10)
```

```python
# итак, у нас теперь есть список топ юзеров
ratings_filtered = grouped_dt[grouped_dt['timestamp'] >= 100]
top_users = ratings_filtered['userId'].tolist()
top_users
```

```python
# вычистили список оценок он ненужных пользователей
ratings_dt = ratings_dt.loc[ratings_dt['userId'].isin(top_users)]
ratings_dt.head(20)
```

```python
# делаем выборку первой и последней по времени оценки каждого пользователя
users_lifetime = ratings_dt.groupby('userId').agg(['min', 'max'])['timestamp'].reset_index()
users_lifetime.head(20)
```

```python
# переделываем в нормальный формат дат
users_lifetime['min']=pd.to_datetime(users_lifetime['min'], unit='s')
users_lifetime['max']=pd.to_datetime(users_lifetime['max'], unit='s')

# не смог придумать лямба функцию, сделал обычную, которая тупо вычитает даты
def minusator(row):
    x = row['max'] - row['min']     
    return x

# прогоняем функцию с помощью апплай
users_lifetime['timelife'] = users_lifetime.apply(minusator, axis=1)
users_lifetime.head(20)
```

```python
# **Задание 4**

# Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия. Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

# При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

# 1. В переменную years запишите список из всех годов с 1950 по 2010.

# 2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
#     - для каждой строки пройдите по всем годам списка years
#     - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
#     - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

# 3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

# 4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга

#начинаем традиционно
import pandas as pd
movies_dt = pd.read_csv('movies.csv')
ratings_dt = pd.read_csv('ratings.csv')

# создаем общую таблицу Х
x = ratings_dt.merge(movies_dt, how='left')

# делаем таблицу индекс- название фильма - средний ретинг
films_rating = x.groupby('title').mean().reset_index()[['title', 'rating']]

films_rating.head()
```

```python
# 1. В переменную years запишите список из всех годов с 1950 по 2010.
years = [str(i+1950) for i in range(61)]
print(years)
```

```python
# 2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
#     - для каждой строки пройдите по всем годам списка years
#     - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
#     - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

# создаем функцию с вариантами оценки
def production_year(row):
    for year in years:
        if year in row['title']:
            return year
    return '1900'
```

```python
# 3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'
films_rating['year'] = films_rating.apply(production_year, axis=1)
films_rating.head(10)
```
